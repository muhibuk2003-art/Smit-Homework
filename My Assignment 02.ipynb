# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks=int(input("Enter Marks: "))
if(marks>=90):
    print("A Grade")
elif(marks>=80):
    print("B Grade")
elif(marks>=70):
    print("C Grade")
elif(marks>=60):
    print("D Grade")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day=int(input("Enter Day Number: "))
if day==1:
    print("Monday")
elif day==2:
    print("Tuesday")
elif day==3:
    print("Wednesday")
elif day==4:
    print("Thursday")
elif day==5:
    print("Friday")
elif day==6:
    print("Saturday")
elif day==7:
    print("Sunday")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month_number=int(input("Enter Month Number: "))
if month_number==12 or month_number==1 or month_number==2:
    print("Winter")
elif month_number==3 or month_number==4 or month_number==5:
    print("Spring")
elif month_number==6 or month_number==7 or month_number==8:
    print("Summer")
elif month_number==9 or month_number==10 or month_number==11:
    print("Autumn")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
op=str(input("Enter Operator: "))
if op=='+':
    print("addition")
if op=='-':
    print("subtraction")
if op=='*':
    print("multiplication")
if op=='/':
    print("division")
else:
    print("Invalid operator")
```
